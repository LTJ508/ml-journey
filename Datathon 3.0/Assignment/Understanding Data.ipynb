```python
df['transaction_date_time'].describe()
```

```python
df['transaction_date_time'] = pd.to_datetime(df['transaction_date_time'], format="%d-%m-%Y %H:%M")
```

```python
df['transaction_date_time'].describe()
```

```python
df.info()
```

```python
df['month'] = df['transaction_date_time'].dt.month

# Grouping by month and calculating count of transactions
grouped = df.groupby('month').size()

# Plotting bar plot of count of transactions by month
grouped.plot(kind='bar')

# Adding labels and title
plt.xlabel('Month')
plt.ylabel('Count of Transactions')
plt.title('Bar plot of Count of Transactions by Month')

# Display the plot
plt.show()
```

```python
df.info()
```

```python
# Grouping by quarter and calculating count of transactions
grouped = df.groupby('quarter').size()

# Plotting bar plot of count of transactions by quarter
grouped.plot(kind='bar')

# Adding labels and title
plt.xlabel('Quarter')
plt.ylabel('Count of Transactions')
plt.title('Bar plot of Count of Transactions by Quarter')

# Display the plot
plt.show()
```

```python
df['year'] = df['transaction_date_time'].dt.year

# Grouping by month and calculating count of transactions
grouped = df.groupby('year').size()

# Plotting bar plot of count of transactions by month
grouped.plot(kind='bar')

# Adding labels and title
plt.xlabel('Year')
plt.ylabel('Count of Transactions')
plt.title('Bar plot of Count of Transactions by Year')

# Display the plot
plt.show()
```

```python
grouped = df.groupby(['year', 'month']).size()

# Calculating average transaction count per month per year
average_transactions = grouped.groupby(level='year').mean()

print(average_transactions)
```
