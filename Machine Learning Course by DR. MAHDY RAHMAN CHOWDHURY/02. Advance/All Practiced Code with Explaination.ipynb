### If it is imbalance then what should we do?

=> first we can delete extra data by **Undersampling**\
undersampling means reducing the major class samples to match minor class samples

```python
from sklearn.utils import resample #importing resample module

majority_class = df6[df6['class'] == 'Positive'] #dividing the datasets
minority_class = df6[df6['class'] == 'Negative'] #dividing the datasets

n_samples = len(minority_class) #count the minor sample
#removing extra sample randomly from major class
majority_downsampled = resample(majority_class, replace = False, n_samples = n_samples, random_state = 42)

#concating two class in one datasets
balanced_df = pd.concat([minority_class, majority_downsampled])
```

```python
value_counts = balanced_df.groupby('class').size().reset_index(name = 'Count') #counting the unique value in class column
value_counts
```

=> now its totally balanced datasets

#### Oversampling(SMOTE): increase the sample of minor class

```python
from imblearn.over_sampling import SMOTE #importing SMOTE
```

```python
df5.head() #already encoded
```

```python
#using df5 as it is encoded
X = df5.drop('class', axis = 1) #drop the class which was our target
X #now X is the datasets of features
```

```python
y = df5['class'] #here y is now the datasets of our target
y
```

```python
smote = SMOTE(random_state = 42) #creating an instance of SMOTE with random state 42
smote
```

```python
X_resampled, y_resampled = smote.fit_resample(X, y) #training the model SMOTE
X_resampled #returned value check
```

```python
y_resampled #retured value check
```

```python
oversampled_df = pd.DataFrame(X_resampled, columns = X.columns) #now making a new dataframe
oversampled_df
```

```python
oversampled_df['class'] = y_resampled #adding the class datasets in the dataframe
oversampled_df
```

```python
value_counts = oversampled_df.groupby('class').size().reset_index(name = 'Count') #counting the unique value in class column
value_counts #now check it is now balanced ny oversampling
```
