### If it is imbalance then what should we do?

=> first we can delete extra data by **Undersampling**\
undersampling means reducing the major class samples to match minor class samples

```python
from sklearn.utils import resample #importing resample module

majority_class = df6[df6['class'] == 'Positive'] #dividing the datasets
minority_class = df6[df6['class'] == 'Negative'] #dividing the datasets

n_samples = len(minority_class) #count the minor sample
#removing extra sample randomly from major class
majority_downsampled = resample(majority_class, replace = False, n_samples = n_samples, random_state = 42)

#concating two class in one datasets
balanced_df = pd.concat([minority_class, majority_downsampled])
```

```python
value_counts = balanced_df.groupby('class').size().reset_index(name = 'Count') #counting the unique value in class column
value_counts
```

=> now its totally balanced datasets
