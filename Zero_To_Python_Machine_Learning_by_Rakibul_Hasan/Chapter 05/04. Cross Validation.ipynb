# Cross data validation, parameter tuning, model selection

# K Fold Cross Validation

```python
#first loading the necessary module and datasets
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

```python
#loading the data and definig the feature and target
iris = load_iris()

X = iris.data
y = iris.target
```

First K_Fold divide the datasets in x partition, then it will chose one partition as test and other partition as training, then it will change the partition for test up to x times, and every time it will check the testing accuracy. then it will find the average of test accuracy

```python
# lets perform the somulation
from sklearn.model_selection import KFold

kf = KFold(n_splits = 5, shuffle = False).split(range(25))

print('{} {:^61}{}'.format('Iteration', 'Training set observations', 'Testing set observation'))

for iteration, data in enumerate(kf, start = 1):
    print('{:^9}{}{:^25}'.format(iteration, data[0], str(data[1])))
```

## we need to use K = 10 as reconmended
## we need to distribute these data in same ratio for each class, this is statify
## cross_val_score do these auto

# lest tune the iris dataset

```python
from sklearn.model_selection import cross_val_score
```

```python
knn = KNeighborsClassifier(n_neighbors = 5)

scores = cross_val_score(knn, X, y, cv = 10, scoring = 'accuracy') #cv = cross validation number
print(scores)
```

```python
print(scores.mean()) #checking the average of scores to check the accuracy of out of sample data
```

# lets try with different value of K = neighbors

```python
k_range = list(range(1, 31))
k_scores = []

for k in k_range:
    knn = KNeighborsClassifier(n_neighbors = k)
    scores = cross_val_score(knn, X, y, cv = 10, scoring = 'accuracy')
    k_scores.append(scores.mean())

print(k_scores)
```

```python
#lets plot these value and observe
import matplotlib.pyplot as plt
%matplotlib inline

plt.plot(k_range, k_scores)
plt.xlabel("Value of K for KNN")
plt.ylabel("Cross-Validated Accuracy")
```

# Lets compare with the logistic regression

```python
knn = KNeighborsClassifier(n_neighbors = 20)
print(cross_val_score(knn, X, y, cv = 10, scoring = 'accuracy').mean())
```

```python
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression(max_iter=1000)
print(cross_val_score(logreg, X, y, cv = 10, scoring = 'accuracy').mean())
```
