# Linear Classification with Feature Scaling

```python
#importing module
from sklearn.datasets import load_iris

iris = load_iris()

X_temp = iris.data
y_temp = iris.target
```

# now we will use two feature instead of 4 only sepal part

## Lets divide first

## Feature scaling
first find the average value of a feture, then subtract the average value from eact data in that feature, then divide these value with the standard deviation\
after that average of each feature will be 0 and standard deviation will be 1

```python
from sklearn.model_selection import train_test_split
from sklearn import preprocessing

#we will take only two features
X, y = X_temp[:, [0, 1]], y_temp

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 33)


#lets standarize these features
scaler = preprocessing.StandardScaler().fit(X_train)

X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

# Lets see what happended here

```python
import numpy as np

print("Training set mean: {:.2f} and Standard deviation: {:.2f}".format(np.average(X_train), np.std(X_train)))

print("Testing set mean: {:.2f} and standard deviation: {:.2f}".format(np.average(X_test), np.std(X_test)))
```

# Lets see the training data in matplotlib

```python
import matplotlib.pyplot as plt

colour_mk = [["red", 's'], ["green", "o"], ["blue", "x"]]
plt.figure("training Data")

for i in range(len(colour_mk)):
    xs = X_train[:, 0][y_train == i]
    ys = X_train[:, 1][y_train == i]
    plt.scatter(xs, ys, c = colour_mk[i][0], marker = colour_mk[i][1])


plt.rcParams['axes.facecolor'] = 'white'
plt.style.use('default')
plt.grid(c = 'gray')

plt.title("Training instances, after scaling")
plt.legend(iris.target_names)
plt.xlabel("Sepal length")
plt.ylabel("Sepal width")
plt.show()
```

# Binary classification => lets predict whatever it is setosa or not => 0 or 1

```python
import copy
y_train_setosa = copy.copy(y_train)
y_train_setosa[y_train_setosa > 0] = 1 #converting 1 and 2 to 1

y_test_setosa = copy.copy(y_test)
y_test_setosa[y_test_setosa > 0] = 1

print("New training target classes:\n{0}".format(y_train_setosa))
```

# SGDClassifier

```python
from sklearn import linear_model 
clf = linear_model.SGDClassifier(loss='log_loss', random_state=42)
```

```python
clf.fit(X_train, y_train_setosa)
```

```python
print(clf.coef_, clf.intercept_)
```

```python
x_min, x_max = X_train[:, 0].min() - .5, X_train[:, 0].max() + .5
y_min, y_max = X_train[:, 1].min() - .5, X_train[:, 1].max() + .5
xs = np.arange(x_min, x_max, 0.5)

fig,axes = plt.subplots()

axes.set_aspect('equal')
axes.set_title('Setosa species classification')
axes.set_xlabel('Sepal length')
axes.set_ylabel('Sepal width')
axes.set_xlim(x_min, x_max)
axes.set_ylim(y_min, y_max)

plt.sca(axes)

plt.scatter(X_train[:, 0][y_train == 0], X_train[:, 1][y_train == 0], c='red', marker='s')
plt.scatter(X_train[:, 0][y_train == 1], X_train[:, 1][y_train == 1], c='black', marker='x')

ys = (-clf.intercept_[0]- xs * clf.coef_[0, 0]) / clf.coef_[0, 1]

plt.plot(xs, ys)
plt.show()
```

```python
print ('If the flower has 4.6 petal width and 3.2 petal length is a {}'.format(iris.target_names[clf.predict(scaler.transform([[4.6, 3.2]]))]))
```
