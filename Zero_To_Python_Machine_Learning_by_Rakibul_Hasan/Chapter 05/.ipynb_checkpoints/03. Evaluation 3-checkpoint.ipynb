# What is neighbors number and in which value it will give better accuracy?

```python
from sklearn.datasets import load_iris
iris = load_iris()

X = iris.data
y = iris.target

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.4, random_state = 4)

from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

## lets run a loop and record the accuracy score using the n_neighbors value from 1 to 25

```python
neighbors_settings = list(range(1, 26)) #defining a list using range function
scores = []

for k in neighbors_settings:
    knn = KNeighborsClassifier(n_neighbors = k)
    knn.fit(X_train, y_train)
    y_pred = knn.predict(X_test)
    scores.append(accuracy_score(y_test, y_pred))
```

# lets plot the scores and k to see visually

```python
import matplotlib.pyplot as plt #importing marplotlib
#for showing inline
%matplotlib inline 


plt.plot(neighbors_settings, scores) #defining axis

plt.xlabel("Value of K for KNN") #labeling the x-axis
plt.ylabel("Testing Accuracy") #labeling the y-axis
```

```python
#checking with 10
knn = KNeighborsClassifier(n_neighbors = 10)
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)
accuracy_score(y_test, y_pred)
```

```python
knn.predict([[3, 5, 4, 2]]) #predicting by out of data example
```

```python
print("Predicted target name: ", iris['target_names'][knn.predict([[3, 5, 4, 2]])])
```

# Now lets check the training and test accuracy

```python
%matplotlib inline

X_train, X_test, y_train, y_test = train_test_split(X, y, stratify = y, random_state = 4) #spliting the data

training_accuracy = [] #definig list for storing training accuracy for each neighbors
test_accuracy = [] #defining list for storing test accuracy

neighbors_settings = range(1, 26) #defining list using range funtion

for n_neighbors in neighbors_settings: #this loop will define knn for each neighbors number from 1 to 25
    knn = KNeighborsClassifier(n_neighbors = n_neighbors) #creating instance of KNeighborsClassifier
    knn.fit(X_train, y_train) #making relation in the model
    training_accuracy.append(knn.score(X_train, y_train)) #checking training accuracy and storing it
    test_accuracy.append(knn.score(X_test, y_test)) #checking test accuracy and storing it

plt.plot(neighbors_settings, training_accuracy, label = "training accuracy") #ploting neighbors vs training accuracy
plt.plot(neighbors_settings, test_accuracy, label = "test accuracy") #ploting neighbors vs test accuracy

plt.ylabel("Accuracy") #defining the label of y axis
plt.xlabel("Number of Neighbors") #defining the label for x axis
plt.legend() #showing the info of data
```

## so, we can see that when neighbors number is 1, training accuracy is 100% and it decreases as the neighbors number increases
## also for test accuracy, decreases as neihbors number increases
