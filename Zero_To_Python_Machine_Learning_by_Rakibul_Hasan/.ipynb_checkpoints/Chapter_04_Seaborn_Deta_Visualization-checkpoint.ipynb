```python
%matplotlib inline
import seaborn as sns
```

```python
iris = sns.load_dataset('iris')
```

```python
iris_numeric = iris.drop(columns=['species'])
```

```python
correlation = iris_numeric.corr()
correlation
```

annot = the value of correlation<br>
cmap = color template

```python
sns.heatmap(iris_numeric.corr(), cmap = 'coolwarm', annot = True)
```

# The sns.lmplot() function combines regplot() and FacetGrid, which enables fitting and plotting a regression model for each subset of the data defined by the hue parameter.

hue = color according to the column given, number of unique value = number of color<br>
palette = set of color, which color set will be use, Set1 defined for unique color set<br>
markers = defined the shape to be used, here, o = circle, s = square, D = diamonds

```python
sns.set(style="whitegrid", color_codes=True) #set style of background
sns.lmplot(x="sepal_length", y="petal_length", hue="species", data = iris, palette="Set1", markers=["o", "s", "D"])
```

hue = color according to the column given, number of unique value = number of color<br>
data = is the dataframe that would be used to plot the graph<br>
style = detarmined the shape, number of unique value = number of shape

```python
sns.scatterplot(x="sepal_length", y = "sepal_width", hue="species", style="species", data=iris)
```

# legend False will hide the information of the plot

```python
sns.lmplot(x="sepal_length", y = "sepal_width", data=iris, hue="species", legend=False)
```

# fit_reg = fit regression, will enable or disable the bestfit line

```python
sns.lmplot(x="sepal_length", y="petal_length", hue="species", data=iris, palette="Set1", fit_reg=False, markers=['o', 's', 'D'])
```

# Here,
height = graph height<br>
vars = list of variable(column) will used for scatter plot

```python
sns.pairplot(iris, height=5, vars=["sepal_length", "petal_length"], markers=["o", "s", "D"], hue="species")
```

# for scatter plot of all possible combination

```python
sns.pairplot(iris, markers=["o", "s", "D"], hue="species")
```

```python
sns.pairplot(iris, markers=["o", "s", "D"], kind="reg", hue="species")
```

orientation h means horizontal, feature will be plot on y axis

```python
sns.boxplot(data=iris, orient = "h")
```

# adding two new column with some calculation

```python
iris["ID"] = iris.index
iris["ratio_sepal"] = iris["sepal_length"] / iris["sepal_width"]
iris["ratio_petal"] = iris["petal_length"] / iris["petal_width"]
```

```python
iris
```

```python
sns.lmplot(x="ID", y="ratio_sepal", data=iris, hue="species", markers=["o", "s", "D"], fit_reg=True, legend=True)
```

```python
sns.lmplot(x="ID", y="ratio_petal", data=iris, hue="species", markers=["o", "s", "D"], fit_reg=True, legend=True)
```

```python
sns.stripplot(data=iris, x="species", y="sepal_length", jitter=False)
```

# jitter, if false, all data will place in the line of categorical variable, x variable

```python
sns.stripplot(data=iris, x="species", y="petal_length", jitter=False)
```

```python
sns.stripplot(data=iris, x="species", y="petal_width", jitter=True)
```

```python
sns.swarmplot(data=iris, x="species", y="petal_length", hue="species", size=4)
```

```python
sns.swarmplot(data=iris, x="species", y="petal_width", hue="species", size=3)
```

```python
sns.swarmplot(data=iris, x="species", y="sepal_length", hue="species", size=4)
```

```python
sns.swarmplot(data=iris, x="species", y="sepal_width", hue="species", size=4)
```
