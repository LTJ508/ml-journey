# 1st Model using KNN

```python
import numpy as np
import pandas as pd
from sklearn import datasets
from sklearn.neighbors import KNeighborsClassifier
```

```python
iris = datasets.load_iris() #importing the dataset in an object
```

```python
#defining feature and target matrix in a variable
X = iris.data
y = iris.target
```

```python
# creating an instance of KNN Classifier
knn = KNeighborsClassifier(n_neighbors = 1)
```

```python
#fitting the model
knn.fit(X, y)
```

```python
#predicitng 
knn.predict([[3, 5, 4, 2]])
```

```python
print("Predicter target name: ", iris['target_names'][knn.predict([[3, 5, 4, 2]])])
```

```python
X_new = np.array([[3, 5, 4, 2]]) #its 2D array thats why two 3rd bracket
print("X_new.shape: ", X_new.shape)
```

```python
mypredict = knn.predict(X_new)
```

```python
mypredict
```

```python
print("Prediction: ", mypredict)
print("Predicted target name: ", iris['target_names'][mypredict])
```

# using two out of data

```python
X_new = [[3, 5, 4, 2], [5, 4, 3, 2]]
knn.predict(X_new)
```

# Changing the nearest neihbors number to 3

```python
knn = KNeighborsClassifier(n_neighbors = 3)
knn.fit(X, y)
```

```python
knn.predict(X_new) #predicitng using previous data
```
