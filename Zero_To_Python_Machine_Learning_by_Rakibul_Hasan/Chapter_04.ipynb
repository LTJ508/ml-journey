# defining dataframe  using data
pandas dataframe has 3 part
1. data = is the main data also known as record
2. index = is the label for the row, first row known as index 0, this is used for searching and retriving data from the dataframe
3. columns = name of column, in machine learning it is known as features and only one from them are our target

```python
import pandas as pd
df = pd.DataFrame(data = data[1:, 1:],
                  index = data[1:, 0],
                  columns = data[0, 1:])
```

```python
%matplotlib inline
import seaborn as sns
```

```python
iris = sns.load_dataset('iris')
```

```python
iris_numeric = iris.drop(columns=['species'])
```

```python
correlation = iris_numeric.corr()
correlation
```

```python
sns.heatmap(iris_numeric.corr(), cmap = 'coolwarm', annot = True)
```

```python
sns.set(style="whitegrid", color_codes=True)
sns.lmplot(x="sepal_length", y="petal_length", hue="species", data = iris, palette="Set1", markers=["o", "s", "D"])
```

```python
sns.scatterplot(x="sepal_length", y = "sepal_width", hue="species", style="species", data=iris)
```
